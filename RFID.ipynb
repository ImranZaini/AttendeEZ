```python
import serial
import time
from datetime import datetime
import firebase_admin
from firebase_admin import credentials, db

# Initialize Firebase with your credentials for Database 1
cred1 = credentials.Certificate("service1.json")
app1 = firebase_admin.initialize_app(cred1, {'databaseURL': 'https://akp505-default-rtdb.asia-southeast1.firebasedatabase.app/'}, name='app1')
ref_students1 = db.reference(app=app1)
# Initialize Firebase with your credentials for Database 2
cred2 = credentials.Certificate("service2.json")
app2 = firebase_admin.initialize_app(cred2, {'databaseURL': 'https://cpt101-f6b5f-default-rtdb.asia-southeast1.firebasedatabase.app/'}, name='app2')
ref_students2 = db.reference(app=app2)

# Open the serial port for communication with Arduino
arduino_port = "COM3"  # Replace 'x' with the actual COM port number
ser = serial.Serial(arduino_port, 9600, timeout=1)

def update_attendance_status(uid, ref_students):
    # Update the "attendance" status to True for the given UID
    ref_attendance = ref_students.child(uid).child('attendance')
    ref_attendance.set(True)


def update_isInside_status(uid, ref_students):
    # Update the 'isInside' status to True
    ref_isInside = ref_students.child(uid).child('isInside')

    # Get the current status
    current_status = ref_isInside.get()

    if current_status is None:
        ref_isInside.set(True)
    else:
        new_status = not current_status
        ref_isInside.set(new_status)
        print(f"Status toggled to {new_status} for UID: {uid}")

def add_uid_to_database1(uid, ref_students1):
    # Add the UID to the 'Students UID' node
    ref_students1.child('Students UID').child(uid).set({
        'attendance': False,
        'isInside': False
    })

try:
    while True:
        try:
            data = ser.readline().decode().strip()
            print("Received data:", data)  # Debug: Print received data

            if data.startswith("Card UID:"):
                uid = data.split(":")[1].strip()
                print("UID:", uid)

                # Check if the UID corresponds to any student in Database 1
                UIDCheck1 = ref_students1.child('Students UID')
                snapshot1 = UIDCheck1.child(uid).get()
                if snapshot1:
                    update_attendance_status(uid, UIDCheck1)
                    update_isInside_status(uid, UIDCheck1)
                else:
                    print("UID not associated with any student in Database 1.")

                # Check if the UID corresponds to any student in Database 2
                UIDCheck2 = ref_students2.child('Students UID')
                snapshot2 = UIDCheck2.child(uid).get()
                if snapshot2:
                    update_attendance_status(uid, UIDCheck2)
                    update_isInside_status(uid, UIDCheck2)
                else:
                    print("UID not associated with any student in Database 2.")

        except UnicodeDecodeError:
            pass
        else:
            time.sleep(0.1)

except KeyboardInterrupt:
    print("Script interrupted.")
```
